```python
slices = [4,1,2,3]

activity = ["Sleeping","Eating", "Working", "Playing"]

cols = ['g','b','r','m']

plt.pie(slices,
        labels=activity,
        colors=cols,
        startangle=45,
        shadow=True,
        explode=(0.2,0.2,0.2,0.3),
        autopct="%1.1f%%")        #      
plt.title('Scatter Plot')
plt.show()
```

```python
df=pd.read_csv("/content/drive/MyDrive/Personal things (important)/iris.csv")
```

```python
fig, ax = plt.subplots()

plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

ax.scatter(df['sepal_length'],df['sepal_width'])

fig.set_size_inches(10,5)
```

```python
colors = {'setosa':'c','versicolor':'m','virginica':'b'}          #Dictionary of color
fig, ax = plt.subplots()

plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

for i in range(len(df['sepal_width'])):
  ax.scatter(df['sepal_length'][i],df['sepal_width'][i],color=colors[df['species'][i]])
```

```python
data = df['species'].value_counts()

name = data.index
count = data.values

plt.title('Scores')
plt.xlabel('Names')
plt.ylabel('Total Count')

plt.bar(name,count)
```

```python
slices =df['species'].value_counts()
activities = ['Setosa','Versicolor','Virginica']
cols = ['g','b','r']

plt.pie(slices,
  labels=activities,
  colors=cols,
  startangle=90,
  shadow= True,
  explode=(0,0.1,0),
  autopct='%1.0f%%',
  radius=1)
 
plt.title('Pie Plot')
plt.show()
```
